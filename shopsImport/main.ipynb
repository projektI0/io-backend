```python
from shopsResponse import elements
```

```python
def getTags(tags):
    return (tags.get('name'), tags.get('addr:street'), tags.get('addr:housenumber'), tags.get('addr:postcode'))
def mapElement2(element):
    return (element['id'], element.get('lat'), element.get('lon'), *getTags(element['tags']))

def filterElement(elementMapped):
    for i in range(6):
        if elementMapped[i] is None:
            return False
    return True
    
commonData = list(filter(filterElement, map(mapElement2, elements['elements'])))
```

```python
with open('shopsOutput.sql', 'w+') as shopsFile:
    shopsFile.write('INSERT INTO SHOP (name, latitude, longitude, address) VALUES\n')
    for (id, lat, lon, name, street, houseNumber, postCode) in commonData:
        correntName = name.replace('\'','\'\'')
        shopsFile.write(f"('{correntName}', {lat}, {lon}, '{postCode} Kraków, {street} {houseNumber}'),\n")
        print(f"({name}, {lat}, {lon}, '{postCode} Kraków, {street} {houseNumber}')")
```
